```python
# b. Write a function called GetBasics that, given the integers p and n, returns a list, say 
# basics, of p different randomly-chosen values from 0, 1, …, n – 1 with the remaining 
# values in 0, 1, …, n – 1 being stored in the list nonbasics. Do this by starting with
# basics = [ ] and nonbaiscs = [0, 1, …, n – 1]. Then sequentially use the “pop” method to 
# move a random element of nonbasics to basics exactly p times. (5 points)
```

```python
import numpy as np
import sys
import random

def GetBasics(p,n):
    nonbasics = [i for i in range(n)]
    basics = []
    ####### randomly-chosen values from 0, 1, …, n – 1
    while p != 0:
        random_num = random.randint(0, n-1)
        print("At present, nonbasics is")
        print(nonbasics)
        print("Pop the number of",random_num,"which is",nonbasics[random_num])
        basics = basics + [nonbasics[random_num]]
        print("At present, basics is")
        print(basics)
        nonbasics.pop(random_num)
        n = n - 1
        p = p - 1
    print("######################################################")
    print("Finally, the result of basics is:")
    print(basics)
    print("Finally, the result of nonbasics is:")
    print(nonbasics)
    return basics,nonbasics
##### Main function:
p = int(input("Please input the number of p:"))
n = int(input("Please input the number of n:"))
while p >= n:
    print("Your p >= n!!!!!!!!! Please re-enter p,n,and keep p < n!")
    p = int(input("Please input the number of p:"))
    n = int(input("Please input the number of n:"))
GetBasics(p,n)

```

```python
# c. Given the lists basics and nonbasics from part (b), write a function called GetBandN that 
# returns the submatrix B consisting of the columns of A indexed by the list basics together 
# with the submatrix N consisting of the remaining columns of A. For example, if n = 5 and 
# the basics are [2, 4] and the nonbasics are [1, 5, 3], then B consists of columns 2 and 4 of 
# A while N consists of columns 1, 5, and 3 of A. (5 points)
```

```python
import numpy as np
import sys
import random

def GetAMatrix(p,n):
    A = np.random.uniform(-10,10,size = (p,n))
    return A

def GetBasics(p,n):
    nonbasics = [i for i in range(n)]
    basics = []
    ####### randomly-chosen values from 0, 1, …, n – 1
    print("######################################################")
    while p != 0:
        random_num = random.randint(0, n-1)
        print("At present, nonbasics is")
        print(nonbasics)
        print("Pop the number of",random_num,"which is",nonbasics[random_num])
        basics = basics + [nonbasics[random_num]]
        print("At present, basics is")
        print(basics)
        nonbasics.pop(random_num)
        n = n - 1
        p = p - 1
    print("######################################################")
    print("Finally, the result of basics is:")
    print(basics)
    print("Finally, the result of nonbasics is:")
    print(nonbasics)
    print("######################################################")
    return basics,nonbasics

def GetBandN(A,basics,nonbasics):
    B = A[:,basics]
    N = A[:,nonbasics]
    return B,N
    
#### Main function 
p = int(input("Please input the number of p:"))
n = int(input("Please input the number of n:"))
while p >= n:
    print("Your p >= n!!!!!!!!! Please re-enter p,n,and keep p < n!")
    p = int(input("Please input the number of p:"))
    n = int(input("Please input the number of n:"))
A = GetAMatrix(p,n)
print("The matrix of A is:")
print(A)
basics,nonbasics = GetBasics(p,n)
B,N = GetBandN(A,basics,nonbasics)
print("B is:")
print(B)
print("N is:")
print(N)
```

```python
import numpy as np
import sys
import random

def GetAMatrix(p,n):
    A = np.random.uniform(-10,10,size = (p,n))
    return A

def GetBasics(p,n):
    nonbasics = [i for i in range(n)]
    basics = []
    ####### randomly-chosen values from 0, 1, …, n – 1
    print("######################################################")
    while p != 0:
        random_num = random.randint(0, n-1)
        print("At present, nonbasics is")
        print(nonbasics)
        print("Pop the number of",random_num,"which is",nonbasics[random_num])
        basics = basics + [nonbasics[random_num]]
        print("At present, basics is")
        print(basics)
        nonbasics.pop(random_num)
        n = n - 1
        p = p - 1
    print("######################################################")
    print("Finally, the result of basics is:")
    print(basics)
    print("Finally, the result of nonbasics is:")
    print(nonbasics)
    print("######################################################")
    return basics,nonbasics

def GetBandN(A,basics,nonbasics):
    B = A[:,basics]
    N = A[:,nonbasics]
    return B,N

def inverse(B):
    Result_List = []
    try:
        inverse_B = np.linalg.inv(B)
        Result_List = Result_List + [True,inverse_B]
        return Result_List
    except:
        Result_List = Result_List + [False]
        return Result_List
        
#### Main function 
p = int(input("Please input the number of p:"))
n = int(input("Please input the number of n:"))
while p >= n:
    print("Your p >= n!!!!!!!!! Please re-enter p,n,and keep p < n!")
    p = int(input("Please input the number of p:"))
    n = int(input("Please input the number of n:"))
A = GetAMatrix(p,n)
print("The matrix of A is:")
print(A)
basics,nonbasics = GetBasics(p,n)
B,N = GetBandN(A,basics,nonbasics)
print("B is:")
print(B)
inverse_B = inverse(B)
print("Inverse of B is:")
if(inverse_B[0] == True):
    print(inverse_B[1])
    print(inverse_B[0])
else:    
    print(inverse_B[0])
    print("B does't have inverse matrix!!!!")
```
